```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load your data from the file into a DataFrame
df = pd.read_csv('/Users/ziyun/Documents/Research/Merge/weathermerged_file_copy.csv')

# Iterate over unique 'sitenum' values and plot data for each sitenum
for sitenum, data in df.groupby('sitenum'):
    plt.figure(figsize=(10, 6))
    plt.plot(data['date'], data['temp'], label='Temperature')
    plt.plot(data['date'], data['maxtemp'], label='Max Temperature')
    plt.plot(data['date'], data['mintemp'], label='Min Temperature')
    plt.xlabel('Date')
    plt.ylabel('Temperature (°C)')
    plt.title(f'Data for Site {sitenum}')
    plt.legend()
    plt.xticks(rotation=45)
    # Calculate date interval based on data range
    date_range = pd.to_datetime(data['date']).max() - pd.to_datetime(data['date']).min()
    if date_range.days > 365:
        date_interval = 'month'
    elif date_range.days > 30:
        date_interval = 'week'
    else:
        date_interval = 'day'
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator(interval_multiples=True))
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
